```python
from ultralytics import YOLO
```

```python
import torch
import torchvision
print(torch.__version__)
print(torchvision.__version__)
print(torch.version.cuda)
device = torch.device('cpu')
print(f'Using device: {device}')
```

```python
model=YOLO("yolov8n.pt")  # pretrained YOLOv8n model
model.to(device)
model.train(data='./car/data.yaml', epochs=10, imgsz=640,batch=16,project='runs/detect',   name='train' )
```

```python
model.val()
model.save('yolov8_car.pt')
```

```python
import matplotlib.pyplot as plt
import os
from PIL import Image
from IPython.display import display

# Function to display an image
confusion_matrix_path = 'runs\detect/train3/confusion_matrix.png'
results_image_path = 'runs\detect/train3/results.png'

def display_image(image_path):
    if os.path.exists(image_path):
        try:
            img = Image.open(image_path)
            display(img)
        except Exception as e:
            print(f"Error loading image {image_path}: {e}")
    else:
        print(f"Image not found: {image_path}")

# Display confusion matrix
display_image(confusion_matrix_path)

# Display results image
display_image(results_image_path)
```
